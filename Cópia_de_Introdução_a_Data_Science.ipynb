```python
contagem_de_lingua = tmdb.original_language.value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ["original_language","total"]
contagem_de_lingua.head()
```

```python
sns.barplot(x="original_language", y="total",data = contagem_de_lingua)
```

```python
sns.catplot(x = "original_language", kind = "count", data = tmdb) # Use the DataFrame variable 'tmdb' instead of the string "tmdb"
```

```python
plt.pie(contagem_de_lingua["total"], labels = contagem_de_lingua["original_language"])
```

```python
total_por_lingua = tmdb["original_language"].value_counts() # Use the correct method 'value_counts()' to count occurrences in a Series.
total_de_ingles = total_por_lingua.loc["en"]
total_geral = total_por_lingua.sum()
total_geral = total_por_lingua.sum()
total_do_resto =  total_geral - total_de_ingles
print(total_de_ingles, total_do_resto)
```

```python
dados = {
    'lingua': ['ingles', 'outros'],
    'total': [total_de_ingles, total_do_resto]
}

dados
```

```python
pd.DataFrame(dados)
```

```python
sns.barplot(x="lingua", y="total", data = dados)
```

```python
plt.pie(dados["total"], labels = dados["lingua"])
```

```python
total_por_lingua_de_outros_fimes = tmdb.query("original_language != 'en'").original_language.value_counts()
```

```python
filmes_sem_lingua_original_ingles = tmdb.query("original_language != 'en'")
plt.figure(figsize=(5,10))
```

```python
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_ingles)
```

```python
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_ingles, aspect = 2)
```

```python
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_ingles, aspect = 2, order = total_por_lingua_de_outros_fimes.index)
```

```python
sns.catplot(x = "original_language", kind = "count", data = filmes_sem_lingua_original_ingles, palette = "rocket", aspect = 2, order = total_por_lingua_de_outros_fimes.index)
```
